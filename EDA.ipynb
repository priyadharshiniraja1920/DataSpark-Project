```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("D:/Mini_Project2/Capsone_Project2/Merged_DataSet.csv")

gender_counts = df['gender'].value_counts()

plt.pie(
    gender_counts, 
    labels=gender_counts.index, 
    autopct='%1.1f%%', 
    colors=["pink", "purple"], 
    startangle=90
)
plt.title("Gender Distribution")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("D:/Mini_Project2/Capsone_Project2/Merged_DataSet.csv")

# Bar Chart: Average Quantity by Gender
avg_quantity_by_gender = df.groupby('gender')['quantity'].mean()
avg_quantity_by_gender.plot(kind='bar', color=['pink', 'brown'], figsize=(6, 5))
plt.title('Average Quantity by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Quantity')
plt.xticks(rotation=0)
plt.show()
```

```python
# Scatter Plot: Quantity vs Unit Price (USD)
plt.figure(figsize=(8, 5))
plt.scatter(df['quantity'], df['unit_price_usd'], alpha=0.5, color='orange')
plt.title('Quantity vs Unit Price (USD)')
plt.xlabel('Quantity')
plt.ylabel('Unit Price (USD)')
plt.grid(True)
plt.show()

```

```python
# Sort the dataset by unit cost in descending order
top_products = df.nlargest(25, 'unit_cost_usd')

# Plot bar chart
plt.figure(figsize=(6, 6))
plt.bar(top_products['product_name'], top_products['unit_cost_usd'], color='pink')

# Adding title and labels
plt.title('Top 25 Products by Unit Cost')
plt.xlabel('Product Name')
plt.ylabel('Unit Cost (USD)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Display the chart
plt.show()
```

```python
# Line Chart
plt.figure(figsize=(25, 6))
plt.plot(top_products['product_name'], top_products['unit_cost_usd'], marker='o', color='green', linestyle='--')

plt.title('Unit Cost of Top 25 Products')
plt.xlabel('Product Name')
plt.ylabel('Unit Cost (USD)')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Histogram
plt.figure(figsize=(10, 6))
plt.hist(df['unit_cost_usd'], bins=20, color='pink', edgecolor='black')

plt.title('Histogram: Distribution of Unit Costs')
plt.xlabel('Unit Cost (USD)')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("D:/Mini_Project2/Capsone_Project2/Merged_DataSet.csv")

# selected columns
columns_to_compare = ['quantity', 'unit_price_usd', 'exchange']
colors = ['blue', 'orange', 'green'] 

plt.figure(figsize=(10, 8))

# Loop through columns and plot each
for i, column in enumerate(columns_to_compare, start=1):
    plt.subplot(len(columns_to_compare), 1, i)  # Create subplots
    plt.hist(df[column], bins=20, color=colors[i - 1], edgecolor='black', alpha=0.7)
    plt.title(f'Histogram: {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
