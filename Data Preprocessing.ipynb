## Importing Libraries & Dataset

```python
import pandas as pd 

customer_df=pd.read_csv('Customers.csv',encoding='latin')
exchange_rate_df=pd.read_csv('Exchange_Rates.csv',encoding='latin')
products_df=pd.read_csv('Products.csv',encoding='latin')
sales_df=pd.read_csv('Sales.csv',encoding='latin')
stores_df=pd.read_csv('Stores.csv',encoding='latin')
```

```python
# need to check for the missing values
print("Missing values in Customers Dataset:\n", customer_df.isnull().sum(), "\n")
print("Missing values in Exchange Rates Dataset:\n", exchange_rate_df.isnull().sum(), "\n")
print("Missing values in Products Dataset:\n", products_df.isnull().sum(), "\n")
print("Missing values in Stores Dataset:\n", stores_df.isnull().sum(), "\n")
print("Missing values in Sales Dataset:\n", sales_df.isnull().sum(), "\n")
```

```python
# to handle the missing values
customer_df.ffill(inplace=True)
exchange_rate_df.ffill(inplace=True)
products_df.ffill(inplace=True)
sales_df.ffill(inplace=True)
stores_df.ffill(inplace=True)
```

```python
# to check for duplicates in the datasets
print("Duplicate rows in Customers Dataset:", customer_df.duplicated().sum())
print("Duplicate rows in Exchange Rates Dataset:", exchange_rate_df.duplicated().sum())
print("Duplicate rows in Products Dataset:", products_df.duplicated().sum())
print("Duplicate rows in Stores Dataset:", stores_df.duplicated().sum())
print("Duplicate rows in Sales Dataset:", sales_df.duplicated().sum())
```

```python
# Removing duplicates in the datasets
customer_df.drop_duplicates(inplace=True)
exchange_rate_df.drop_duplicates(inplace=True)
products_df.drop_duplicates(inplace=True)
stores_df.drop_duplicates(inplace=True)
sales_df.drop_duplicates(inplace=True)
```

```python

# Filter out rows where 'Zip Code' contains non-numeric values
customer_df = customer_df[customer_df['Zip Code'].str.isnumeric()]
# Convert 'Zip Code' to integer
customer_df['Zip Code'] = customer_df['Zip Code'].astype('int32')
```

```python
customer_df['Birthday'] = pd.to_datetime(customer_df['Birthday'], errors='coerce', format='%m-%d-%Y')
customer_df['Birthday'] = customer_df['Birthday'].fillna(pd.to_datetime(customer_df['Birthday'], format='%m/%d/%Y', errors='coerce'))
```

```python
customer_df.info()
```

```python
customer_df.to_csv('cleaned_customer_df.csv', index=False)
```

```python

# Extract numeric part
products_df['Unit Cost USD'] = products_df['Unit Cost USD'].str.extract('(\d+)')
# Convert to integer, handle NaN by dropping them or filling them
products_df['Unit Cost USD'] = products_df['Unit Cost USD'].dropna().astype('int32')
```

```python
# Extract numeric part
products_df['Unit Price USD'] = products_df['Unit Price USD'].str.extract('(\d+)')
# Convert to integer, h-andle NaN by dropping them or filling them
products_df['Unit Price USD'] = products_df['Unit Price USD'].dropna().astype('int32')
```
